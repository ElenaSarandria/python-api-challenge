```python
# Dependencies and Setup
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import requests
import gmaps
import gmaps.datasets
import os
import json
# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_data_to_load = "output_data/weather_data.csv"
weather_data = pd.read_csv(weather_data_to_load)
dropna_weather_data = weather_data.dropna()
dropna_weather_data.index = np.arange(1, len(dropna_weather_data)+1)
del dropna_weather_data["Unnamed: 0"]
dropna_weather_data.head(20)
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
# Locations
locations = dropna_weather_data[["Lat", "Lng"]]
humidity = dropna_weather_data["Humidity"]
# Plot Heatmap
fig = gmaps.figure()
# Createlayer
heatmap_layer = gmaps.heatmap_layer(locations, weights=humidity)
fig.add_layer(heatmap_layer)
heatmap_layer.max_intensity = 80
heatmap_layer.point_radius = 5
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Filter vacation with cloudiness below 30
vacation_cloud = dropna_weather_data[dropna_weather_data["Cloudiness"] < 30]
# Filter vacation with max temp above 60 degrees F
vacation_above_60_degrees = vacation_cloud[vacation_cloud["Max Temp"] > 60]
# Filter vacation with max temp below 100 degrees F
vacation_below_100_degrees = vacation_above_60_degrees[vacation_above_60_degrees["Max Temp"] < 100]
# Filter vacation with wind speed below 15 mph
vacation_slow_wind = vacation_below_100_degrees[vacation_below_100_degrees["Wind Speed"] < 15]
# Filter vacation with humidity below 50 %
perfect_vacation = vacation_slow_wind[vacation_slow_wind["Humidity"] < 50]
city_perfect_vacation = perfect_vacation.reset_index()
del city_perfect_vacation["index"]
city_perfect_vacation
```

```python
gmaps.configure(api_key=g_key)
vacation_locations = city_perfect_vacation[["Lat", "Lng"]]
vacation_humidity = city_perfect_vacation["Humidity"]
vacation_fig = gmaps.figure()
vacation_layer = gmaps.heatmap_layer(vacation_locations, weights=vacation_humidity, dissipating=False, 
                                     max_intensity=50, point_radius=2.5)
vacation_fig.add_layer(vacation_layer)
vacation_fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
gmaps.configure(api_key=g_key)
hotels = []
# Loop 
for city in range(len(city_perfect_vacation["City"])):
    lat = city_perfect_vacation.loc[city]["Lat"]
    #perfect_vacation["Lat"].iloc[0]
    #perfect_vacation.reset_index(drop=True)
    
    lng = city_perfect_vacation.loc[city]["Lng"]
    #perfect_vacation["Lng"].iloc[0]
    #perfect_vacation.reset_index(drop=True)
    
    city_coords = f"{lat},{lng}"
    params = {"location": city_coords, 
              "types": "lodging",
              "radius": 5000,
              "key": g_key
             }
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"   
    hotel_request = requests.get(base_url, params=params)
    hotel_response = hotel_request.json()
    try:
        hotels.append(hotel_response["results"][0]["name"])
    except:
        hotels.append("Nearest hotel not found")

# Dataframe with nearest hotel
city_perfect_vacation["Hotel Name"] = hotels
city_perfect_vacation
    
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in city_perfect_vacation.iterrows()]
locations = city_perfect_vacation[["Lat", "Lng"]]

# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content= [f"Hotel Name: {hotel}" for hotel in hotels])
vacation_fig.add_layer(markers)

# Display Map
vacation_fig
```
